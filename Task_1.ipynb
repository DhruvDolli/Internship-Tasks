```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading dataset

file_path = r"C:\Users\dhruv\Internship\Task_1\student\student-mat.csv"
dataset = pd.read_csv(file_path)
```

```python
# First 5 result of the dataset

dataset.head()
```

```python
# Information about the data

print(dataset.isnull().sum())
print(dataset.dtypes)
print(dataset.shape)
```

```python
# To remove duplicates from the dataset

dataset = dataset.drop_duplicates()
```

```python
# To improve the readability of the dataset 

dataset = pd.read_csv(file_path, sep=";")
dataset.columns
```

```python
# Data Analysis
```

```python
# 1 

avg_score = dataset['G3'].mean()
print(f"The Average Score in Math in G3 is: {avg_score} ")
```

```python
# 2

students_above_15 = dataset[dataset['G3'] > 15].shape[0]
print(f"Students that scored above 15 in their final grade are: {students_above_15}")
```

```python
# 3

correlation = dataset['studytime'].corr(dataset['G3'])
print(f"The correlation between study time and G3 is: {correlation}")
```

```python
# 4

avg_grade_gender = dataset.groupby("sex")["G3"].mean()
print(f"The higher average grade by gender are {avg_grade_gender}")
```

```python
# Data Visualization
```

```python
# 1

plt.figure(figsize=(8, 5))
plt.hist(dataset["G3"], bins=10, color="green", edgecolor="black", alpha=0.7)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades (G3)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# 2

plt.figure(figsize=(8, 5))
plt.scatter(dataset["studytime"], dataset["G3"], color="blue", alpha=0.6)
plt.xlabel("Study Time (hours)")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade (G3)")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
# 3

plt.figure(figsize=(6, 4))
plt.bar(avg_grade_gender.index, avg_grade_gender.values, color=["pink", "blue"], alpha=0.7, edgecolor="black")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.title("Average Final Grade by Gender")
plt.ylim(0, max(avg_grade_gender) + 2)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
