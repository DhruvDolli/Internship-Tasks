```python
sp_data['Close Lag'] = sp_data['Close'].shift(1)
sp_data['Mean Moving Avg'] = sp_data['Close'].rolling(window = 3).mean()
print(sp_data)
```

```python
from statsmodels.tsa.arima.model import ARIMA
import pmdarima as pm
```

```python
sp_data.set_index('Date', inplace = True)
series = sp_data['Close']
```

```python
series.plot(title = 'Time Series Data')
plt.show()
```

```python
sp_data['Close'].plot(title = 'Stock Price', figsize = (12, 6))
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid()
plt.show()
```

```python
auto_model_1 = pm.auto_arima(sp_data['Close'], seasonal=False, stepwise=True, trace=True)
print(auto_model_1.summary())
```

```python
order = auto_model_1.order

model_1 = ARIMA(sp_data['Close'], order=order)
model_fit_1 = model_1.fit()
print(model_fit_1.summary())
```

```python
# Forecast next 30 days
forecast_steps = 30
forecast1 = model_fit_1.forecast(steps=forecast_steps)

# Plot
plt.figure(figsize=(12, 5))
plt.plot(sp_data['Close'], label='Historical')
plt.plot(pd.date_range(start=sp_data.index[-1], periods=forecast_steps+1, freq='B')[1:], forecast1, 
         label='Forecast', color='red')
plt.title("ARIMA Stock Price Forecast (Next 30 Business Days)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()
```

```python
train_size = int(len(sp_data) * 0.8)
train, test = sp_data['Close'][:train_size], sp_data['Close'][train_size:]
```

```python
log_data = np.log(train)

diff_data = log_data.diff().dropna()
```

```python
auto_model = pm.auto_arima(train, seasonal=False, stepwise=True, suppress_warnings=True)
order = auto_model.order

model = ARIMA(train, order = order)
model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps = len(test))

forecast_df = pd.DataFrame({'Actual': test, 'Predicted': forecast}, index = test.index)

forecast_df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Actual Prices', color='blue')
plt.plot(test.index, forecast, label='Predicted Prices', color='red')
plt.title('Actual vs Predicted Stock Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

print(f"📉 MAE  : {mae:.2f}")
print(f"📉 RMSE : {rmse:.2f}")
```
