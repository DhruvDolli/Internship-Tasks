```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\dhruv\Internship\Task_8\student_performance_dataset\exams.csv"
data = pd.read_csv(file_path)
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
features = ['gender', 'race/ethnicity', 'test preparation course', 'math score', 'reading score', 
            'writing score']
new_data = data[features]
new_data
```

```python
new_data['total_score'] = new_data['math score'] + new_data['reading score'] + new_data['writing score']
new_data
```

```python
new_data['percentage'] = new_data['total_score'] / 300 * 100
new_data
```

```python
def result(row):
    if row['percentage'] > 40:
        return "1"
    else:
        return "0"

new_data['Result'] = new_data.apply(result, axis = 1)
new_data
```

```python
new_data['gender'] = new_data['gender'].map({'male': 0, 'female': 1})
new_data
```

```python
new_data['race/ethnicity'] = new_data['race/ethnicity'].map({'group A': 0, 'group B': 1, 'group C' : 2, 
                                                            'group D' : 3, 'group E' : 4})
new_data
```

```python
new_data['test preparation course'] = new_data['test preparation course'].map({'none': 0, 'completed': 1})
new_data
```

```python
new_data['test preparation course'].unique()
```

```python
new_data['Result'].value_counts()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = new_data[['gender', 'race/ethnicity', 'test preparation course']]
y = new_data['Result']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=36)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
param_grid = {
    'C' : [0.01, 0.1, 1, 10],
    'penalty' : ['l1', 'l2'],
    'solver' : ['liblinear']
}
```

```python
grid_search = GridSearchCV(model, param_grid, cv = 5, scoring = 'accuracy', verbose=1)
grid_search.fit(X_train_scaled, y_train)
```

```python
print("Best Parameter:", grid_search.best_params_)
print("Best Cross Validation:", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_scaled)
```

```python
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:", classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\dhruv\Internship\Task_8\fraud_detection_data_1.csv"
data = pd.read_csv(file_path)
data
```

```python
data['newbalanceDest'].value_counts()
```

```python
drop_features = ['isFlaggedFraud', 'step']
fd = data.drop(drop_features, axis=1)
fd
```

```python
fd.isnull().sum()
```

```python
fd['type'].unique()

# Convert type into numbers
# TRANSFER - 0, PAYMENT - 1, CASH OUT - 2, CASH IN -3, DEBIT -4
```

```python
fd['type'] = fd['type'].map({'TRANSFER': 0, 'PAYMENT': 1, 'CASH_OUT' : 2, 
                                                            'CASH_IN' : 3, 'DEBIT' : 4})
fd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

for col in ['nameOrig', 'nameDest', 'transaction_id']:
    fd[col] = le.fit_transform(fd[col])

fd
```

```python
X = fd.drop('isFraud', axis=1)
y = fd['isFraud'] 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=36)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=36)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test) 
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(dt, filled=True, feature_names=X.columns, class_names=['Not Fraud', 'Fraud'])
plt.show()
```
